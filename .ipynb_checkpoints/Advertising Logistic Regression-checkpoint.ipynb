# Advertising Logistic Regression #
In this project, we'll aim to create a logistic regression model which will predict whether or not a user will click on an add based on the features of that user.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
from scipy.stats import stats
sns.set_style("darkgrid")
```

```python
ad_data = pd.read_csv("advertising.csv")
ad_data.head()
```

```python
ad_data.info()
```

```python
ad_data.describe()
```

### Exploratory Data Analysis ###

```python
ad_data["Age"].plot(kind="hist", bins=30)
plt.xlabel("Age")
plt.show()
```

Most users tend to be in their thirties, there are also a couple of spikes in the late twenties and early forties.

```python
sns.jointplot(x="Age",y="Area Income",data=ad_data).annotate(stats.pearsonr)
```

There seems to be very little correlation between age and area income

```python
sns.jointplot(x="Age",y="Daily Time Spent on Site",data=ad_data,kind="kde",color="red").annotate(stats.pearsonr)
```

Younger people (especially 30 year olds) tend to spent the most amount of time on the site

```python
sns.jointplot(x="Daily Time Spent on Site",y="Daily Internet Usage",color="green",data=ad_data).annotate(stats.pearsonr)
```

There's a positive association between daily internet usage and time spent on the site, but not a very strong positive correlation

```python
sns.pairplot(ad_data,hue="Clicked on Ad",palette="coolwarm")
```

### Building the logistic regression model ###
Begin by doing a train-test split on the data

```python
from sklearn.model_selection import train_test_split
```

```python
ad_data.columns
```

```python
X = ad_data[['Daily Time Spent on Site', 'Age', 'Area Income', 'Daily Internet Usage', 'Male']]
y = ad_data["Clicked on Ad"]
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30, random_state=101)
```

### Training the model ###
Train the model and get prediction values for the testing data

```python
from sklearn.linear_model import LogisticRegression
```

```python
logmodel = LogisticRegression()
logmodel.fit(X_train,y_train)
```

```python
predictions = logmodel.predict(X_test)
```

### Evaluating Model Performance ###

```python
from sklearn.metrics import classification_report, confusion_matrix
print(confusion_matrix(y_test,predictions))
print("\n")
print(classification_report(y_test,predictions))
```
